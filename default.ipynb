```python
from sklearn.linear_model import LinearRegression
import numpy as np
import  matplotlib.pyplot as plt
```

```python
data = np.loadtxt('LinearRegression/ex1data1.txt', delimiter=',')
lr = LinearRegression()
lr.fit(np.expand_dims(data[:,0],axis=1),np.expand_dims(data[:,1],axis=1))
sk_theta = [lr.intercept_[0],lr.coef_[0, 0]]
x = np.linspace(data[:, 0].min(), data[:, 0].max(), 100)
f1 = sk_theta[0] + (sk_theta[1] * x)
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(x, f1, 'y', label='Prediction')
ax.scatter(data[:,0], data[:,1], label='Data Set', s=10)
ax.legend(loc=2)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Linear Regression')
plt.show()
sk_theta
```

```python
from LinearRegression.LinearRegression import square_loss

plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']  # 
plt.figure(dpi=120,figsize=(12,8))

fig = plt.figure()
ax3d = fig.add_subplot(projection='3d')  # 3d\
theta0 = np.linspace(-4, 4, num=256)
theta1 = np.linspace(-4, 4, num=256)
xx, yy = np.meshgrid(theta0, theta1)

x = np.expand_dims(data[:,0],axis=1)
y = np.expand_dims(data[:,1],axis=1)
z=np.zeros([256,256])
for i in range(len(theta0)):
    for j in range(len(theta1)):
        pred = theta0[i]+theta1[j]*x
        z[i, j] = square_loss(pred,y)/(2*pred.shape[0])

ax3d.plot_surface(xx, yy, z, alpha=0.8, rstride=1, cstride=1, cmap="rainbow")

plt.tight_layout()
plt.show()
z
```
