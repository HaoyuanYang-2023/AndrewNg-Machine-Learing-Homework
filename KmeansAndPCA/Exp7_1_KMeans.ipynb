```python
image_data = loadmat('data/bird_small.mat')
image = image_data.get('A')
image.shape
```

```python
plt.figure()
plt.imshow(image)
plt.axis("off")
plt.title("Origin Image")
plt.show()
```

```python
image = image / 255.

# reshape the array
image_x = np.reshape(image, (image.shape[0] * image.shape[1], image.shape[2]))
image_x.shape
```

```python
clf_image = KMeans(k=16, max_iter=1000)
cls, centroids = clf_image.fit(image_x)
```

```python
print(cls.size, centroids.size)
```

```python
x_recovered = centroids[cls.astype(int),:]
x_recovered = np.reshape(x_recovered,  (image.shape[0], image.shape[1], image.shape[2]))
```

```python
plt.imshow(x_recovered)
plt.axis("off")
plt.title("Compressed Image")
plt.show()
```
